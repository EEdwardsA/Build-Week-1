```
# Encode string values as numeric. Put "not sure" somewhat arbitrarily between 
# "I do not think this is a priority, but should receive some resources" and
# "This cause deserves significant resources but less than the top priorities"
# (This played into later not wanting to ultimately use data that corresponded to all possible answers)

# In order: 
# 0 = I do not think any resources should be devoted to this cause  
# 1 = I do not think this is a priority, but should receive some resources
# 2 = Not considered / Not sure 
# 3 = This cause deserves significant resources but less than the top priorities 
# 4 = This cause should be a near-top priority        
# 5 = This cause should be the top priority<e5><ca>(Please choose one)
ea2018_num = ea2018.replace({'I do not think any resources should be devoted to this cause':0, 
                             'I do not think this is a priority, but should receive some resources':1,
                             'Not considered / Not sure':2,
                             'This cause deserves significant resources but less than the top priorities':3,
                             'This cause should be a near-top priority':4,
                             'This cause should be the top priority<e5><ca>(Please choose one)':5})
ea2018_num.head()
```

```
# Find null values

ea2018_num.isnull().sum()
```

```
# Drop null values and find out how many rows I'm left with 
# (make sure final number of rows is at least equal to 5 times number of columns)

ea2018_nonull = ea2018_num.dropna()
ea2018_nonull.shape
```
