```
# maybe I'm thinking about this wrong. What are better ways to get at answers?
# I could be dropping all the 2 values from cause prioritzation colums (representing unsure)
# I could separate observations into groups based on which cause column they put a 5 in (or 4/5 in)
# and then compare their personality trait scores to the rest of the population

# Let's do animal welfare first
top_animal = ea2018_nonull[ea2018_nonull['cause_import_animal_welfare'] == 5]
top_animal
```

```
not_top_animal = ea2018_nonull[ea2018_nonull['cause_import_animal_welfare'] != 5]
not_top_animal
```

```
# Do these two groups have different personality trait scores?
# First, run 1-sample t-test

ea2018_nonull.mean()
```

```
from scipy import stats
```

```
# People whose top priority is animal welfare, conscientiousness against population conscientiousness
stats.ttest_1samp(top_animal['big_five_conscientious_scale'], 9.986521)
```

```
# People whose top priority is animal welfare, differ on extraversion from population?
stats.ttest_1samp(top_animal['big_five_extraverted_scale'], 7.574558)
```

```
# Animal lovers agreeableness
stats.ttest_1samp(top_animal['big_five_agreeable_scale'], 9.098568)
```

```
# People whose top priority is animal welfare, differ on emotional stability from population?
### THIS ONE IS POTENTIALLY INTERESTING

stats.ttest_1samp(top_animal['big_five_emotionally_stable_scale'], 9.785173)
```

```
stats.ttest_1samp(top_animal['big_five_open_scale'], 10.994945)
```

```
# What are some other ways I could measure this? I could compare people who gave animal welfare a 
# 5 with people who gave animal welfare a 0
hate_animals = ea2018_nonull[ea2018_nonull['cause_import_animal_welfare'] == 0]

stats.ttest_ind(hate_animals, top_animal)
```

```
# Compare the groups on extroversion

stats.ttest_ind(hate_animals['big_five_extraverted_scale'], top_animal['big_five_extraverted_scale'])
```

```
stats.ttest_ind(hate_animals['big_five_conscientious_scale'], top_animal['big_five_conscientious_scale'])
```

```
stats.ttest_ind(hate_animals['big_five_agreeable_scale'], top_animal['big_five_agreeable_scale'])
```

```
# Again, this one is pretty interesting
# Emotional stability/neuroticism is highly predictive of assigning either a 0 or a 5 to 
# animal welfare, whereas (below) openness is the trait on which 0-assigners and 5 assigners 
# *differ* *the* *least*

stats.ttest_ind(hate_animals['big_five_emotionally_stable_scale'], top_animal['big_five_emotionally_stable_scale'])
```

```
stats.ttest_ind(hate_animals['big_five_open_scale'], top_animal['big_five_open_scale'])
```
